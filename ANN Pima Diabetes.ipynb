```python
#Create first network with Keras
import keras
from keras.models import Sequential
from keras.layers import Dense###will connect all the neurons 
import numpy
import pandas as pd
import sklearn
from sklearn.preprocessing import StandardScaler

```

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)

# load pima indians dataset
#dataset = numpy.loadtxt('pima-indians-diabetes.csv', delimiter=",")
dataset = pd.read_csv('pima-indians-diabetes.csv',header=None,)##stops the first line from being taken as the header line

#data=pd.DataFrame(dataset) #data is panda but dataset is something else
dataset.head()

```

```python
dataset.info()
```

```python
# split into input (X ie dependent variables) and output (Y ie independent variables) variables
X = dataset.iloc[:,0:8]   #0-8 columns are dependent variables - remember 8th column is not included
Y = dataset.iloc[:,8]#8 column is independent variable
#Y= Y.reshape(-1,1) ##can't take dataframes it has to be in a list format
```

```python
##iloc must be used to get the dataframe in a list format

```

```python
type(X)
```

```python
X.shape
```

```python
scaler = StandardScaler()###standard scaler is added because the data is in plenty of different scales because it won't work properly with all the different scales 
X = scaler.fit_transform(X)
```

```python
# create model
model = Sequential()###this creates the neural network architecture 
# model.add((Interconnection)Dense(1000, input_dim=8, activation='relu')) # 1000 neurons
# model.add(Dense(100, activation='tanh')) # 100 neurons with tanh activation function
model.add(Dense(500, input_dim=8, activation='relu')) # 500 neurons
# 93.36% accuracy with 500 neurons
# 86.99% accuracy with 100 neurons
# 85.2% accuracy with 50 neurons
# 81.38% accuracy with 10 neurons
model.add(Dense(1, activation='sigmoid')) # 1 output neuron

##input 8 is the number of independent variables## 

##model.add(Dense(500, activation='relu')) another feature has been added just remove input dim

##there's no standard calculation you have to have high understanding of data but hyperparamters can be done to an extent
```

```python
##next is what to use to look at the loss
##
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Fit the model
model.fit(X, Y, epochs=150, batch_size=10,  verbose=2) # 150 epoch, 10 batch size, verbose = 2
##the epochs is how many times you run all the batches through the neural network - you can do this multiple times until the loss is very low
##verbose gives information
##batches is 10 rows being trained at a time
###loss is binary because we have a binary neural dependent variable
```

```python
##you can see the loss is very low and accuracy is at 90%
##the evaluation will show what this means in terms of an accuracy score--> its like a summary of all the data points
```

```python
# evaluate the model
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

```

```python
# calculate predictions
predictions = model.predict(X)    # predicting Y only using X
print(predictions) ###it has to be rounded because sigmoid is giving data between 0 and 1 but we either 0 or 1 
```

```python
# Round predictions
#rounded = [int(numpy.round(x, 0)) for x in predictions]
rounded = [round(x[0]) for x in predictions]
print(rounded)

print("Rounded type: ", type(rounded)) # rounded is a 'list' class
print("Shape of rounded: ", len(rounded))
print("Dataset type: ", type(dataset)) # numpy array?
print("Shape of dataset: ", dataset.shape)
```

```python
###confusion matrix to display how well the classification worked 
import matplotlib.pyplot as plt
import numpy
from sklearn import metrics

actual = Y
predicted = rounded

confusion_matrix = metrics.confusion_matrix(actual, predicted)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show() 
```
