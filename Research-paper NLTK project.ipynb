```python
from nltk.tokenize import word_tokenize
```

```python
from nltk.collocations import QuadgramCollocationFinder
```

```python
import PyPDF2
import nltk
```

```python
stopwords = nltk.corpus.stopwords.words("english") ##corpus is a book in this case its a book of all stopwords words
```

```python
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page in range(len(reader.pages)):
            text += reader.pages[page].extract_text()
        return text

pdf_path = 'JohnnyandAmber.pdf'
pdf_text = extract_text_from_pdf(pdf_path)
```

```python
words = pdf_text
```

```python
words
```

```python
tokens = word_tokenize(pdf_text)
```

```python
filtered_tokens = [word.lower() for word in tokens if word.lower() not in stopwords]
```

```python
filtered_tokens
```

```python
#tk_words = filtered_tokens
```

```python
##NLTK text has to be done to make it possible to get FD 

tk_words = nltk.Text(filtered_tokens) 
```

```python
fd =nltk.FreqDist(tk_words)
```

```python
fd.most_common(20)
```

```python
fd.tabulate(20)
```

```python
fd["hit"]
```

```python
tk_words.concordance("hit") 
```

```python
finder = nltk.collocations.TrigramCollocationFinder.from_words(tk_words)
```

```python
finder.ngram_fd.most_common(5) 
```

```python
finder2 = nltk.collocations.QuadgramCollocationFinder.from_words(tk_words)
```

```python
finder2.ngram_fd.most_common(20) 
```
